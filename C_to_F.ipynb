```
import pandas as pd
import keras
from keras import backend as K
from keras.layers import Activation,Dense
from keras.models import Sequential
from keras.optimizers import Adam
from keras.losses import mean_squared_error
from keras.metrics import *
import numpy as np
```

```
celi=[]
fhar=[]
for i in range(1000):
  c=np.random.randint(0,100)
  celi.append(c)
  fhar.append(c*1.8+32)
  print(fhar[i])
```

```
cel    = np.array(celi,  dtype=float)
far = np.array(fhar,  dtype=float)
cel.shape
```

```
model = Sequential()
model.add(Dense(1,input_shape=[1]))
model.summary()
```

```
model.compile(Adam(lr=0.01),loss = "mean_squared_error" )
```

```
history = model.fit(cel,far,epochs=2000)
print("Training is finshies")
```
